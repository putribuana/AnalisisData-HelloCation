##asal_instansi

```python
import re

def filter_feature(feature):
    # Case folding
    feature = feature.str.lower()

    # Remove punctuation
    feature = feature.str.replace(r'[^\w\s]', '')

    return feature

feature = mentorship['asal_instansi']
filter_feature = filter_feature(feature)
print(filter_feature[1:])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

mentorship["asal_instansi"] = mentorship["asal_instansi"]

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(mentorship["asal_instansi"]))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

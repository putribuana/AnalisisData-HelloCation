```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_gksu = "./datacleaning_gksu.csv"
dataset_gksg = "./datacleaning_gksg.csv"

gks_u = pd.read_csv(dataset_gksu)
gks_g = pd.read_csv(dataset_gksg)
```

```python
gks_u.info()
```

```python
gks_g.info()
```

```python
#tambah kolom "apply_study dengan semua isinya adalah S2"
gks_g["apply_study"] = "S2"
```

```python
#hapus kolom
gks_u = gks_u.drop(["beasiswa_pilihan", "info_dari", "univ_tujuan", "tanggal_transfer"], axis=1)
```

```python
gks_g.info()
```

```python
gks_u = gks_u.dropna()

gks_u.isna().sum()
```

```python
gks_g = gks_g.dropna()

gks_g.isna().sum()
```

##Merged Dataset

```python
mentorship = pd.concat([gks_u, gks_g], ignore_index=True)

mentorship.info()
```

```python
mentorship.describe(include = 'all')
```

```python
# Ubah tipe data kolom "tanggal" menjadi time
mentorship["timestamp"] = pd.to_datetime(mentorship["timestamp"], format="%Y-%m-%d %H:%M:%S")
mentorship.timestamp.value_counts().sort_index()
```

```python
#ubah semua awalaan nama menjadi huruf kapital
mentorship["nama_lengkap"] = mentorship["nama_lengkap"].str.title()

mentorship.nama_lengkap.value_counts().sort_index()
```

```python
mentorship["asal_instansi"] = mentorship["asal_instansi"].str.title()

mentorship["asal_instansi"] = mentorship["asal_instansi"].replace(0, "tidak diketahui")
mentorship.asal_instansi.value_counts().sort_index()
```

```python
mentorship["jurusan"] = mentorship["jurusan"].str.title()

mentorship["jurusan"] = mentorship["jurusan"].replace(0, "tidak diketahui")
mentorship.jurusan.value_counts().sort_index()
```

```python
mentorship["domisili"] = mentorship["domisili"].str.title()

mentorship.domisili.value_counts().sort_index()
```

```python
mentorship["umur"] = mentorship["umur"].astype(int)
mentorship.umur.value_counts().sort_index()
```

```python
mentorship.info()
```

```python
del mentorship["nomor_hp"]
```

```python
del mentorship["nama_ig"]
mentorship.info()
```

```python
mentorship["alasan_join"] = mentorship["alasan_join"].str.title()

mentorship.alasan_join.value_counts().sort_index()
```

```python
mentorship.pilih_kelas.value_counts().sort_index()
```

```python
mentorship.apply_study.value_counts().sort_index()
```

```python
mentorship.to_csv('datamerged_gku_g.csv', index=False)
```

#Visualization and Clustering Process

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

mentorship["alasan_join"] = mentorship["alasan_join"]

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(mentorship["alasan_join"]))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#Cluster

```python
# Membuat scatter plot untuk setiap cluster
for i in range(3):
    mentorship_cluster = mentorship[mentorship["cluster"] == i]
    plt.scatter(mentorship_cluster["jurusan"], mentorship_cluster["apply_study"], label=f"Cluster {i}")

# Menambahkan label dan judul
plt.xlabel("Jurusan")
plt.ylabel("Apply Study")
plt.title("Visualisasi K-Means Clustering")
plt.legend()
plt.show()
```
